```python
unique_apps=[]
duplicate_apps=[]
for app in android:
    name=app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)
print('Number of unique android apps',len(unique_apps))
print('Number of duplicate android apps',len(duplicate_apps))
```

The criterion for removing duplicate rows from the dataset is:
Here I am taking the concern of the reviews columns for all the rows with duplicates being varied by review factor. after observing i can see that data is recorded at different times of the year. Removing duplicates is based on reviews

```python
reviews_max={}
for app in android:
    name=app[0]
    n_reviews=float(app[3])
    #Updating the value in the dictionary with highest review
    if name in reviews_max and reviews_max[name]<n_reviews:
        reviews_max[name]=n_reviews
    #Inserting the new data into the dictionary
    if name not in reviews_max:
        reviews_max[name]=n_reviews
print(len(reviews_max))
#print(reviews_max)
```

From above:
We are finding the max reviews by particular app. Creating a dictionary with that unique app which having highest reviews.

```python
android_clean=[]
already_added=[]
for app in android:
    name=app[0]
    n_reviews=float(app[3])
    if n_reviews==reviews_max[name] and name not in already_added:
        android_clean.append(app)
        already_added.append(name)
print(len(android_clean))
```

Creating a list which has no duplicates in it.

```python
def english_app(string):
    count=0
    for character in string:
        if ord(character)>127 :
            count+=1
    if count<=3:
        return True
    else:
        return False
```

# Validation strategy for an app

- Building a minimal android version of the app, and add it to Google Play
- If the app has a good response from users, we develop it furthur.
- If the app is profitable after six months, we build an IOS version of the app and add it to the app store.
- finally we want to add app on both google and app stores

```python
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage 
    
    return table_percentages


def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
genres_ios=freq_table(free_ios_apps,-5)
for genre in genres_ios:
    total=0
    len_genre=0
    for app in free_ios_apps:
        genre_app=app[-5]
        if genre_app==genre:
            n_ratings=float(app[5])
            total+=n_ratings
            len_genre+=1
    avg_rating=total/len_genre
    print(genre,':',avg_rating)
```
